## 变量的命名

重点: 变量命名规则： 

  1. 变量名的长度不受限制，但其中的字符必须是字母、数字、或者下划线_,（也可以使用中文），而不能使用空格、连字符、标点符号、引号或其他字符。
  2. 变量名的第一个字符不能是数字，而必须是字母或下划线。
  3. Python区分大小写。
  4. 不能将Python关键字用作变量名。

变量严格意义上应该成为"名字"
快递员去快递一样,内存就是一个巨大的货物架,在Python中定义变量,定义变量就相当于给快递贴标签.

### 中间分隔符：sep=" "

打印多个值,用sep设置中间分隔符, 默认为空格

````python
print('数据分析', '就业班 ','python')
````

````python
print('数据分析', '就业班 ','python', sep='--->')
````

### 结尾换行符:end='' "

如果你打印多次，不想一行行打印出来，想都打印在一行，可以在print()后面添加end=""。  
比如：

````python
默认情况下,一条print语句输出后会直接换行,如果想一次输出多个内容,而且不换行,怎么办?
````

````python
a = 1
b = 2
print(a,b)
````

````python
print('同学们，打起精神来')
print('我们马上下课了')
````

````python
print('同学们，打起精神来',end = "")
print('我们马上下课了')
````
